```python
import pandas as pd

data = pd.read_csv("enjoysport.csv").values.tolist()
h = ['Ø'] * (len(data[0]) - 1)

for row in data:
    if str(row[-1]).strip().lower() in ['yes', '1']:   # handles Yes/No, yes/no, 1/0
        for i in range(len(h)):
            if h[i] == 'Ø': h[i] = row[i]
            elif h[i] != row[i]: h[i] = '?'

print("Final Hypothesis:", h)
```
