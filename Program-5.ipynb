```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Step 1: Load dataset
data = pd.read_csv("enjoysport.csv")  # change file name if needed

# Step 2: Encode categorical values
for col in data.columns:
    data[col] = LabelEncoder().fit_transform(data[col])

# Step 3: Split into features (X) and labels (y)
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Step 4: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Step 5: Train model
model = GaussianNB()
model.fit(X_train, y_train)

# Step 6: Predict
y_pred = model.predict(X_test)

# Step 7: Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of Naive Bayes Classifier: {:.2f}%".format(accuracy * 100))

# Step 8: Clean output
print("\nActual Values:   ", list(map(int, y_test)))
print("Predicted Values:", list(map(int, y_pred)))
```
